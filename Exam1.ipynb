```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "dataset01_eurusd4h.csv"
df = pd.read_csv(file_path)

rsi_columns = ['rsi1', 'rsi2', 'rsi3', 'rsi4', 'rsi5', 'rsi6']
df['avg_rsi'] = df[rsi_columns].mean(axis=1)

df['avg_price'] = df[['close1', 'close2', 'close3', 'close4', 'close5', 'close6']].mean(axis=1)

window_size = 10
df['volatility'] = df['avg_price'].rolling(window=window_size).std()

stoch_columns = ['stoch1', 'stoch2', 'stoch3', 'stoch4', 'stoch5', 'stoch6']
df['avg_stoch'] = df[stoch_columns].mean(axis=1)

ema20_slope_columns = ['ema20Slope1', 'ema20Slope2', 'ema20Slope3', 'ema20Slope4', 'ema20Slope5', 'ema20Slope6']
df['avg_ema20_slope'] = df[ema20_slope_columns].mean(axis=1)

bb_up_columns = ['BB_up_percen1', 'BB_up_percen2', 'BB_up_percen3', 'BB_up_percen4', 'BB_up_percen5', 'BB_up_percen6']
df['avg_bb_up'] = df[bb_up_columns].mean(axis=1)

cci_columns = ['cci1', 'cci2', 'cci3', 'cci4', 'cci5', 'cci6']
df['avg_cci'] = df[cci_columns].mean(axis=1)

mom_columns = ['mom1', 'mom2', 'mom3', 'mom4', 'mom5', 'mom6']
df['avg_mom'] = df[mom_columns].mean(axis=1)


plt.figure(figsize=(12, 6))
plt.plot(df['avg_price'], label='Середня ціна', color='tab:red')
plt.xlabel('Час')
plt.ylabel('Середня ціна')
plt.title('Графік середньої ціни акцій')
plt.legend()
plt.show()

selected_rsi = 'rsi1'
plt.figure(figsize=(12, 6))
plt.plot(df[selected_rsi], label=selected_rsi, color='tab:blue')
plt.xlabel('Час')
plt.ylabel(selected_rsi)
plt.title(f'Графік {selected_rsi} акцій')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(df['avg_rsi'], label='Середнє RSI', color='tab:green')
plt.xlabel('Час')
plt.ylabel('Середнє RSI')
plt.title('Графік середнього RSI акцій')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(df['avg_stoch'], label='Середнє Stochastic Oscillator', color='tab:orange')
plt.xlabel('Час')
plt.ylabel('Середнє Stochastic Oscillator')
plt.title('Графік середнього Stochastic Oscillator акцій')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(df['avg_ema20_slope'], label='Середнє EMA Slope', color='tab:purple')
plt.xlabel('Час')
plt.ylabel('Середнє EMA Slope')
plt.title('Графік середнього EMA Slope акцій')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(df['volatility'], label='Стандартне відхилення цін(волатильність)', color='tab:green')
plt.xlabel('Час')
plt.ylabel('Стандартне відхилення цін')
plt.title('Графік стандартного відхилення цін')
plt.legend()
plt.show()
```
